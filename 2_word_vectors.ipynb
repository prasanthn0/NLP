```python
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
train_x = ["I love the book", "This is a great book" , "The fit is great", "I love the shoes"]

class Category:
    BOOKS = "BOOKS"
    CLOTHING = "CLOTHING"
    
train_y = [Category.BOOKS , Category.BOOKS , Category.CLOTHING , Category.CLOTHING]
```

```python
docs = [nlp(text) for text in train_x]
print(docs)
print(docs[0].vector)
```

```python
train_x_vectors = [docs[i].vector for i in range(len(docs))]
#train_x_vectors = [x.vector for x in docs]
```

```python
#we will use a classifier on this now
from sklearn import svm

svm_clf = svm.SVC(kernel = 'linear')
svm_clf.fit(train_x_vectors, train_y)
```

```python
test_x = ['i like the books' , "i like the story" , "i like the shoe's fit in this world"]
test_docs = [nlp(text) for text in test_x]
test_x_vectors = [x.vector for x in test_docs]

svm_clf.predict(test_x_vectors)
```
